Daten laden

```python
file_path = "drive/MyDrive/FINAL_2024.02.05_Preischart_nach_Kalenderwoche.xlsx"
```

```python
df = pd.read_excel(file_path, na_values=[" "])
#Data Frame erstellen, leere Zeilen durch NaN ersetzen.
```

```python
df.head()
```

```python
preise = pd.DataFrame(df)
preise.set_index('KW', inplace=True)
# "KW" als Index festlegen.

teuerste_kw = preise.idxmax()
günstigste_kw = preise.idxmin()
# Für jede Kategorie die Kalenderwoche herauspicken, in der die Preise am höchsten bzw. am tiefsten sind.

höchster_preis = preise.max()
niedrigster_preis = preise.min()
# Höchste und niedrigste Preise heraussuchen.

resultat_höchste = pd.DataFrame({'Höchster Preis': höchster_preis, 'Kalenderwoche': teuerste_kw})
resultat_tiefste = pd.DataFrame({'Tiefster Preis': niedrigster_preis, 'Kalenderwoche': günstigste_kw})
# DataFrames erstellen mit den höchsten und tiefsten Preisen und der dazugehörigen Kalenderwoche.

print("Höchste Preise:")
print(resultat_höchste)
print("\nTiefste Preise:")
print(resultat_tiefste)
```

Tabellen mergen, um damit danach in Datawrapper weiterzuarbeiten

```python
resultat_höchste.index.name = 'Produkt'
resultat_tiefste.index.name = 'Produkt'
# Index auf 'Produkt' umstellen in den DataFrames resultat_höchste und resultat_tiefste

kombinierte_tabelle = pd.merge(resultat_höchste, resultat_tiefste, on='Produkt', suffixes=('_höchste', '_tiefste'))
# DataFrames zusammenfügen

kombinierte_tabelle.reset_index(inplace=True)
# Index resetten um 'Produkt' als reguläre Spalte zu erhalten

kombinierte_tabelle.rename(columns={'Woche_höchste': 'Woche_höchste_Preise', 'Woche_tiefste': 'Woche_tiefste_Preise'}, inplace=True)
# Spalten neu benennen

print("Kombinierte Tabelle:")
print(kombinierte_tabelle)
```

```python
kombinierte_tabelle.to_excel("drive/MyDrive/Tiefste_und_höchste_Preise_nach_KW.xlsx", index=True)
```

```python
# Prozentuale Veränderung berechnen
kombinierte_tabelle['Percentage Change'] = ((kombinierte_tabelle['Höchster Preis'] - kombinierte_tabelle['Tiefster Preis']) / kombinierte_tabelle['Tiefster Preis']) * 100

# DataFrame mit neuer Spalte anzeigen
print(kombinierte_tabelle)
```

Tabellarische Darstellung

```python
# Grösse festlegen
plt.figure(figsize=(10, 6))

# Daten der ersten Kategorie (Auberginen) festlegen
category_data = kombinierte_tabelle.loc[0]

# Daten auslesen
category = category_data['Produkt']
höchster_preis = category_data['Höchster Preis']
tiefster_preis = category_data['Tiefster Preis']
week_highest = category_data['Kalenderwoche_höchste']
week_lowest = category_data['Kalenderwoche_tiefste']
percent_difference = ((höchster_preis - tiefster_preis) / tiefster_preis) * 100

# Plotten
plt.bar(['Tiefster Preis', 'Höchster Preis'], [tiefster_preis, höchster_preis], color=['seagreen', 'indianred'])
plt.title(f'Höchst- und Tiefstpreise für {category}')
plt.xlabel(' ')
plt.ylabel('Preis')
plt.text(0, tiefster_preis, f'Woche {int(week_lowest)}', ha='center', va='bottom', color='seagreen')
plt.text(1, höchster_preis, f'Woche {int(week_highest)}', ha='center', va='bottom', color='indianred')
plt.text(0.5, höchster_preis * 0.8, f'Prozentuale Veränderung: {percent_difference:.2f}%', ha='center', va='bottom')

# Plot anzeigen
plt.tight_layout()
plt.show()
```

Funktion erstellen, um Plot für jede Kategorie schnell anzuzeigen

```python
def plot_product_prices(kombinierte_tabelle, product_name_column, product_name):
    product_data = kombinierte_tabelle[kombinierte_tabelle[product_name_column] == product_name]

    if len(product_data) == 0:
        print(f"No data found for {product_name}.")
        return

    highest_price = product_data['Höchster Preis'].values[0]
    highest_week = product_data['Kalenderwoche_höchste'].values[0]
    lowest_price = product_data['Tiefster Preis'].values[0]
    lowest_week = product_data['Kalenderwoche_tiefste'].values[0]
    percentage_change = product_data['Percentage Change'].values[0]

    plt.bar(['Highest Price', 'Lowest Price'], [highest_price, lowest_price], color=['indianred', 'seagreen'])
    plt.xlabel('Price Type')
    plt.ylabel('Price')
    plt.title(f'{product_name} Prices')
    plt.text(0, highest_price + 0.1, f'Week: {highest_week}\nPrice: {highest_price}', ha='center')
    plt.text(1, lowest_price + 0.1, f'Week: {lowest_week}\nPrice: {lowest_price}', ha='center')
    plt.ylim(0, max(highest_price, lowest_price) + 1)

    plt.text(0.5, (highest_price + lowest_price) / 2, f'Percentage Change: {percentage_change:.2f}%', ha='center')

    plt.show()
```

```python
plot_product_prices(kombinierte_tabelle, 'Produkt', 'Brombeeren')
```

```python
plot_product_prices(kombinierte_tabelle, 'Produkt', 'Spinat')
```

**Um die Darstellung zu erreichen, die ich mir vorstelle für den Plot Chart, muss ich nun die Tabelle nach Höchst- und Tiefstpreisen gruppieren**

```python
lowest_prices = kombinierte_tabelle[['Produkt', 'Tiefster Preis', 'Kalenderwoche_tiefste']].rename(columns={'Tiefster Preis': 'Price', 'Kalenderwoche_tiefste': 'Calendar Week'})
lowest_prices['Price Type'] = 'Lowest Price'

highest_prices = kombinierte_tabelle[['Produkt', 'Höchster Preis', 'Kalenderwoche_höchste']].rename(columns={'Höchster Preis': 'Price', 'Kalenderwoche_höchste': 'Calendar Week'})
highest_prices['Price Type'] = 'Highest Price'

final_df = pd.concat([lowest_prices, highest_prices])

print(final_df)
```
