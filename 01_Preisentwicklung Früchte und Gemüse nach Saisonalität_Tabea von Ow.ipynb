**Durchschnitt ausrechnen pro Kalenderwoche**



```python
df_wochenpreis = df.groupby(df.KW).mean()
# Nach Monat gruppieren und Durchschnitt ausrechnen pro Monat
print(df_wochenpreis)
```

```python
columns_to_delete = ['J', 'M']
# Im neuen DF die überflüssigen Spalten "J" und "M" löschen.

df_wochenpreis = df_wochenpreis.drop(columns=columns_to_delete)

# Display the updated DataFrame
print(df_wochenpreis)
```

```python
df_wochenpreis
```

```python
df_wochenpreis.to_excel('2024.02.05_Preischart_nach_Kalenderwoche.xlsx', index=True)
#Data Frame abspeichern als xlsx, um es danach in Datawrapper weiter zu verwenden.
#Index=True nutzen, um den Index (in diesem Fall die KW) mitabzuspeichern.
```

**Preis-Chart nach Monat**

```python
df_monatspreis = df.groupby(df.M).mean()
# Nach Monat gruppieren und Durchschnitt ausrechnen pro Monat
#df_monatspreis.set_index('KW', inplace=True)4

print(df_monatspreis)
```

```python
df_monatspreis
```

```python
columns_to_delete2 = ['J', 'KW']
# Nun will ich im neuen DF die beiden überflüssigen Spalten "J" und "KW" löschen.

df_monatspreis = df_monatspreis.drop(columns=columns_to_delete2)

# Display the updated DataFrame
print(df_monatspreis)
```

```python
df_monatspreis
```

```python
df_monatspreis.to_excel('2024.02.05_Preischart_nach_Kalendermonat.xlsx', index=True)
#Data Frame abspeichern als xlsx, um es danach in Datawrapper weiter zu verwenden.
#Index=True nutzen, um den Index (in diesem Fall den Monat) mitabzuspeichern.
```

```python
fig, axs = plt.subplots(nrows=len(df_monatspreis.columns), figsize=(10, 4 * len(df_monatspreis.columns)))

for i, column in enumerate(df_monatspreis.columns):
    axs[i].plot(df_monatspreis.index, df_monatspreis[column], marker='o')
    axs[i].set_title(column)
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Average Cost per Kilo')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
#Plotten, um anzuschauen, wie die Tabellen aussehen.
```
